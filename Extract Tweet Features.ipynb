```python
#Code to extract number of hashtags, mentions and media in each tweet
import csv
import pandas as pd
import string
import nltk
from nltk.tokenize import WhitespaceTokenizer
import re
```

```python
#Read the file containing only tweets
tweets_df = pd.read_csv('C:\Users\Sohan\Masters\Maryland iSchool\Courses\Fall 2017\INFM 750 Advanced Data Science\Project\Data\Preliminary data sets\only_tweets.csv')
tweets_df
```

```python
#Count the number of hashtags in each tweet and add a new column with the count
tweets_df['number_of_hashtags'] = tweets_df[['text']].applymap(lambda x: str.count(x, '#'))
tweets_df
```

```python
#Count the number of mentions in each tweet and add a new column with the count
tweets_df['number_of_mentions'] = tweets_df[['text']].applymap(lambda x: str.count(x, '@'))
tweets_df
```

```python
#Count the number of times a media item is shared through each tweet and add a new column with the count
tweets_df['number_of_media'] = tweets_df[['text']].applymap(lambda x: str.count(x, 'http://')) + tweets_df[['text']].applymap(lambda x: str.count(x, 'https://'))
tweets_df
```

```python
#Remove URLs and emojis from tweet text

#Remove URLs from tweets
def remove_URLs(text):
    result = re.sub(r"http\S+", "", text)
    result = re.sub(r"https\S+", "", text)
    return result

#Remove emojis from tweets
def remove_emoji(text):
    text = text.decode('utf-8')
    emoji_pattern = re.compile(
    u"(\ud83d[\ude00-\ude4f])|"  # emoticons
    u"(\ud83c[\udf00-\uffff])|"  # symbols & pictographs (1 of 2)
    u"(\ud83d[\u0000-\uddff])|"  # symbols & pictographs (2 of 2)
    u"(\ud83d[\ude80-\udeff])|"  # transport & map symbols
    u"(\ud83c[\udde0-\uddff])"  # flags (iOS)
    "+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text) # no emoji

tweets_df["text"] = tweets_df["text"].apply(remove_URLs)
tweets_df["text"] = tweets_df["text"].apply(remove_emoji)
tweets_df
```

```python
# Create lexicons
# Add a column to the dataframe with each lexicon category and populate the columns with the values for each tweet
from empath import Empath
lexicon = Empath()

achievement_lst = [ ]
anger_lst = [ ]
emotional_lst = [ ]
envy_lst = [ ]
hate_lst = [ ]
irritability_lst = [ ]
joy_lst = [ ]
love_lst = [ ]
rage_lst = [ ]
sympathy_lst = [ ]

family_lst = [ ]

fashion_lst = [ ]
health_lst = [ ]
exercise_lst = [ ]

music_lst = [ ]

politics_lst = [ ]
government_lst = [ ]

positive_emotion_lst = [ ]
negative_emotion_lst = [ ]

sports_lst = [ ]
traveling_lst = [ ]
vacation_lst = [ ]

for sentence in tweets_df.text:
    kk =  lexicon.analyze(sentence, categories = ['achievement','anger','emotional','envy','hate','irritability','joy','love','rage','sympathy','family','fashion','health','exercise','music','politics','government','positive_emotion','negative_emotion', 'sports', 'traveling','vacation'])
    achievement_lst.append(kk['achievement'])
    anger_lst.append(kk['anger'])
    emotional_lst.append(kk['emotional'])
    envy_lst.append(kk['envy'])
    hate_lst.append(kk['hate'])
    irritability_lst.append(kk['irritability'])
    joy_lst.append(kk['joy'])
    love_lst.append(kk['love'])
    rage_lst.append(kk['rage'])
    sympathy_lst.append(kk['sympathy'])
    family_lst.append(kk['family'])
    fashion_lst.append(kk['fashion'])
    health_lst.append(kk['health'])
    exercise_lst.append(kk['exercise'])
    music_lst.append(kk['music'])
    politics_lst.append(kk['politics'])
    government_lst.append(kk['government'])
    positive_emotion_lst.append(kk['positive_emotion'])
    negative_emotion_lst.append(kk['negative_emotion'])
    sports_lst.append(kk['sports'])
    traveling_lst.append(kk['traveling'])
    vacation_lst.append(kk['vacation'])
    
tweets_df["liwc_achievement"] = achievement_lst
tweets_df["liwc_anger"] = anger_lst
tweets_df["liwc_emotional"] = emotional_lst
tweets_df["liwc_envy"] = envy_lst
tweets_df["liwc_hate"] = hate_lst
tweets_df["liwc_irritability"] = irritability_lst
tweets_df["liwc_joy"] = joy_lst
tweets_df["liwc_love"] = love_lst
tweets_df["liwc_rage"] = rage_lst
tweets_df["liwc_sympathy"] = sympathy_lst
tweets_df["liwc_family"] = family_lst
tweets_df["liwc_fashion"] = fashion_lst
tweets_df["liwc_health"] = health_lst
tweets_df["liwc_exercise"] = exercise_lst
tweets_df["liwc_music"] = music_lst
tweets_df["liwc_politics"] = politics_lst
tweets_df["liwc_government"] = government_lst
tweets_df["liwc_positive_emotion"] = positive_emotion_lst
tweets_df["liwc_negative_emotion"] = negative_emotion_lst
tweets_df["liwc_sports"] = sports_lst
tweets_df["liwc_traveling"] = traveling_lst
tweets_df["liwc_vacation"] = vacation_lst

tweets_df
```

```python
#NLP

#Function used for stemming
porter = nltk.PorterStemmer()

count_error_tok=0
count_total_tok=0

def remove_punctuation(text):
    #Grab the list of standard punctuation symbols that are provided in the string library
    punctuations = string.punctuation # includes following characters: !"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~

    #Don't strip out apostrophes
    excluded_punctuations = ["'"]
    for p in punctuations:
        if p not in excluded_punctuations:
            #Replace each punctuation symbol by a space
            text = text.replace(p, " ") 
    return text

#Tokenizing
#Create a tokenizer from NLTK which will create tokens based on the whitespace in between words
tokenizer = WhitespaceTokenizer()

#Change to lowercase, remove punctuation, remove extra spaces, tokenize, obtain stemmed version of the word
def normalize_review_text(text):
    global count_total_tok
    global count_error_tok
    text = text.lower()
    text = remove_punctuation(text)
    text = " ".join(text.split())
    text_tokens = tokenizer.tokenize(text)
    stemmed_tokens=[]
    for w in text_tokens:
        count_total_tok+=1
        try:
            stemmed_tokens.append(porter.stem(w))
        except:
            stemmed_tokens.append("Error due to utf-8")
            count_error_tok+= 1   
    return text_tokens

tweets_df["text_tokens"] = tweets_df["text"].apply(normalize_review_text)
print "Total count of tokens:",count_total_tok
print "Number of erroneous tokens:",count_error_tok
```

```python
#Remove stopwords
#nltk.download()
from nltk.corpus import stopwords
enStop = stopwords.words('english')
esStop = stopwords.words('spanish')
stopword_list = enStop+esStop+["rt", "@", ":", "co", "amp", "&amp;", "...", "\n", "\r", "#"]

def remove_stopwords(tweets):
    tweets = [w for w in tweets if w not in stopword_list]
    return tweets

tweets_df["text_tokens"] = tweets_df["text_tokens"].apply(remove_stopwords)
tweets_df
```

```python
tweets_df.drop('text', axis=1, inplace=True)
tweets_df.drop('text_tokens', axis=1, inplace=True)
tweets_df
```
